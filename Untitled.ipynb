# **Global Air Quality Index: Analyzing Pollution Patterns Across the World**

## **Business Understanding**

### **Overview**
Air quality is an increasingly vital global issue affecting public health, environmental sustainability, and overall quality of life. Various pollutants such as PM2.5, CO, NO2, and Ozone contribute to the decline in air quality, especially in urban areas where industrialization and vehicular emissions are rampant. This dataset provides a snapshot of air quality measurements across cities worldwide, allowing us to assess the severity of pollution, identify patterns, and propose actionable solutions.

Our goal is to analyze this data to better understand how air pollution varies by location and pollutant type, helping stakeholders like governments, researchers, and the public to make informed decisions for reducing air pollution and protecting human health.

### **Problem Statement**
Air pollution remains one of the most significant environmental and public health challenges globally. With rising industrial activities and increased urbanization, maintaining good air quality has become difficult in many cities. Poor air quality can lead to severe health problems such as respiratory diseases, cardiovascular conditions, and premature death. Understanding the extent of air pollution and its distribution across different regions is essential to developing policies and interventions aimed at improving air quality.

By analyzing air quality data, we seek to determine which cities are most affected by pollution, the main pollutants contributing to these levels, and the impact of geographic and environmental factors on air quality.

---

## **Objectives**

### **Main Objective**
The primary goal of this analysis is to assess the air quality across various global cities and derive actionable insights from the concentrations of different pollutants, with the aim of identifying trends, pollution hotspots, and providing data-driven recommendations for mitigating pollution.

### **Specific Objectives**
1. **Evaluate AQI Levels**: Analyze the overall Air Quality Index (AQI) values and categories across different cities to determine the most and least polluted areas.
2. **Analyze Pollutant-Specific Data**: Investigate the AQI values of specific pollutants (CO, Ozone, NO2, PM2.5) to understand their individual contributions to air pollution.
3. **Categorize Air Quality**: Classify the cities based on AQI categories (e.g., Good, Moderate, Unhealthy) and analyze the distribution of these categories globally.
4. **Perform Geospatial Analysis**: Use geographical coordinates (latitude and longitude) to visualize the spread of air quality levels across the globe, identifying pollution hotspots.
5. **Identify Trends**: Explore patterns over time and across regions to detect trends in pollution levels, determining which regions are improving and which are deteriorating.

## **Data Understanding**

### **Dataset Overview**
The dataset contains detailed air quality data from cities around the world, measuring multiple pollutants and their corresponding Air Quality Index (AQI) values. This data allows us to assess overall air quality, as well as the specific impact of individual pollutants like CO, Ozone, NO2, and PM2.5.

Each row in the dataset represents air quality readings from a specific city on a given day. AQI categories classify these readings into health impact categories (e.g., 'Good', 'Moderate', 'Unhealthy'), providing a holistic view of the air quality in each region.

 ## EDA

### Univariate

```python
#1. Numerical Columns (e.g., AQI Values, lat, lng)
# For each numerical column, we can look at distributions and summary statistics.
# Summary Statistics:
```

```python
data[['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value', 'lat', 'lng']].describe()
```

```python
# List of columns to plot
columns = ['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']

# Create subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Flatten the axs array for easy iteration
axs = axs.ravel()

# Loop over each column and create histograms with Matplotlib
for i, col in enumerate(columns):
    axs[i].hist(data[col].dropna(), bins=30)  # Using .dropna() to avoid NaN issues
    axs[i].set_title(col)

plt.tight_layout()
plt.show()
```

The graphs represent the distribution of values across different features or variables from a dataset. Here's a general explanation of their appearance:

1. **Skewness**: All five histograms are heavily skewed to the right. This means that most of the data points are concentrated towards the lower values, while a few outliers stretch the distribution towards the higher end. Such skewness is common with variables that cannot go below zero but have no clear upper bound.

2. **Peaks and Frequency**: Each histogram has a distinct peak (the mode), which shows the range where most data points fall. After this peak, the frequencies of occurrences rapidly decline as values increase, suggesting that the majority of the data is concentrated in the lower ranges.

3. **Similar Patterns**: While the specific ranges and scales may differ across the graphs, they all share a similar pattern of right-skewed distribution. This consistency could imply that the features being represented have similar data characteristics.

4. **Potential Outliers**: The long tails in each histogram hint at the presence of potential outliers or extreme values that are significantly larger than the bulk of the data. These outliers could disproportionately affect statistical analyses, so it's important to handle them appropriately, possibly by applying transformations or using robust statistics.

In conclusion, these histograms show right-skewed distributions with most data concentrated at lower values and long tails extending to the right, indicating the presence of high-value outliers.

```python
# Numerical vs. Categorical
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='AQI Category', y='AQI Value', data=data)
plt.title('AQI Value across AQI Categories')
plt.show()
```

```python
# Compare averages of numerical variables grouped by categories
```

```python
plt.figure(figsize=(10, 6))
data.groupby('AQI Category')['AQI Value'].mean().plot(kind='bar')
plt.title('Mean AQI Value by AQI Category')
plt.show()
```

This bar chart shows varying levels of categories, with one category significantly higher than the others. The tallest bar indicates a peak or dominant category, while the remaining bars display smaller, yet differing values. This suggests a distribution where one aspect strongly stands out compared to the rest, which may require focused attention or analysis.

```python
#  Geospatial Analysis (lat, lng)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='lng', y='lat', hue='AQI Category', data=data)
plt.title('Geospatial Plot of AQI Categories')
plt.show()
```

This graph displays a global map illustrating various Air Quality Index (AQI) categories represented by color-coded dots. The AQI is a measure of air pollution levels and its potential impact on human health. The legend on the right explains the categories:

- **Blue (Moderate)**: AQI levels indicate that air quality is generally acceptable, but there may be a concern for a very small number of sensitive individuals.
- **Green (Good)**: Air quality is considered satisfactory, and air pollution poses little or no risk.
- **Red (Unhealthy)**: Everyone may experience health effects; members of sensitive groups may experience more serious health effects.
- **Purple (Very Unhealthy)**: Health alertâeveryone may experience more serious health effects.
- **Orange (Unhealthy for Sensitive Groups)**: Sensitive groups may experience health effects, but the general public is less likely to be affected.
- **Maroon (Hazardous)**: Health warnings of emergency conditions. The entire population is more likely to be affected.

The map shows that air quality varies across different regions globally. Areas with higher concentrations of orange and red dots indicate regions with more significant air quality concerns, while blue and green dots suggest regions with generally good to moderate air quality.

Key observations include:
- A large number of regions in Africa, parts of South America, and South Asia show moderate AQI levels.
- Regions in the United States, Europe, and parts of China display a mix of good and moderate air quality.
- Some clusters, particularly in Southeast Asia, show more concerning levels of unhealthy or very unhealthy air quality.

This visualization helps identify global trends in air quality and highlights areas where air pollution is a more severe concern.
